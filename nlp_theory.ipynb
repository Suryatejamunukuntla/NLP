#### Word Embedding 

In Natural language processing (NLP), word embelling is a term used for the <u>representation of words</u> for the text analysis typically in the form of <u>real valued vector</u> that encoders the meaning of word such that the words that are closure in the vector space are expected to be similar in meaning. 

It is classified into two types :

    1. Count of Frequency

    2. Deep Learning (Word2Vec -> CBOW,Skipgram)

#### Word2Vec (Feature Representation)

It is a techique for natural language processing,pulished in 2013 by Google. The Word2Vec algorithm is used in natural neural network model to learn word association from a large corpus/text.Once trained,such a model can detect similar words or suggest additional words for a particular sentence.As the name suggests,Word2Vec represents each distinct word with a particular list of number called a vector.
